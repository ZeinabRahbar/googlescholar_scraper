```python
import requests
import time
from bs4 import BeautifulSoup

# Define the search query
query = "graph nwtwork hyperspectral image classification"

# Scrape paper data from the first 20 pages
for page in range(6):
    # Calculate the start parameter for each page
    start = page * 10

    # Send an HTTP GET request to Google Scholar with the year restriction
    url = f"https://scholar.google.com/scholar?q={query}&as_ylo=2020&start={start}"
    response = requests.get(url)

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.text, "html.parser")

    # Find all paper elements
    papers = soup.find_all("div", class_="gs_ri")

    # Iterate over the papers and print the title and authors
    for paper in papers:
        title = paper.find("h3", class_="gs_rt").text.strip()
        authors = paper.find("div", class_="gs_a").text.strip()
        print("Title:", title)
        print("Authors:", authors)
        print("--------------------")
    time.sleep(2)
```
